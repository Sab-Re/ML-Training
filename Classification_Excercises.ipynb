To make sure the algorithms can process them, we need to change our binary ("yes" or "no") columns into numeric values:

```python
# Lists columns with yes/no 
binary_cols = ['default', 'housing']

# Change columns 'default' and 'housing' to numeric values
for col in binary_cols:
    if col in df.columns:
        df[col] = df[col].map({'yes': 1, 'no': 0})
```

```python
df.head(5)
```

We need to convert non-binary columns such as Job or Marital Status to numeric values as well.

Now, we could do the same as with binary columns and e.g. assign each marital status a different number -- but think about this a moment from a computer's perspective: When Married is replaced with 0, Single with 1 and Divorced with 2, that would imply a numeric relation between Single and Married.

In other terms, Divorced would be 2x Single - which not only sounds strange, but leads to very weird results as well.

So instead, we follow an approach called **One-Hot Encoding**: Each marital status get a new column, e.g. marital_single, marital_married and marital_divorced, where 0 or 1 indicates the status:

```python
# Encode the more complex categorical columns
df = pd.get_dummies(data=df, columns = ['job', 'marital', 'education' , 'month'], \
                                   prefix = ['job', 'marital', 'education' , 'month'])
df.head()
```

<h1><center>4. TRAIN & EVALUATE YOUR MODEL(S)</center></h1>

```python
# To proceed the training, datasets needs to be splited to train and testing samples.
X = df.drop(columns=['y'])
y = df['y']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

<h3><center>Decision Tree</center></h3>

```python
# Train a Decision Tree model
dectree = DecisionTreeClassifier()
dectree.fit(X_train, y_train)
y_pred = dectree.predict(X_test)
```

```python
# Get the accuracy score for the Decision Tree model
acc = dectree.score(X_test, y_test)
print(f'Accuracy score: {acc*100}%')
```
